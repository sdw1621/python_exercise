###**12-4 선 그래프의 범례**

```python
import matplotlib.pyplot as plt
from matplotlib import rc

xdata = ['안지영','홍지수','황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
plt.legend(loc='upper center')

plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

###**12-5 그래프 선 스타일 설정**

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

xdata = ['안지영','홍지수','황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('세 명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

###**12-6 X/Y축의 범위와 눈금 설정**

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
  y = 2*x
  ydata.append(y)

plt.plot(xdata, ydata, label='y=2x')
plt.xlim(0,35)
plt.ylim(0,70)
plt.xticks(list(range(0,35,2)))
plt.yticks(list(range(0,70,5)))

plt.grid(True)
plt.show()
```

###**12-9 파이 그래프 그리기**

```python
import matplotlib.pyplot as plt
from matplotlib import rc

pets = ['개','고양이','기타','기르지않음']
portion = [18,3.4,3.1,75.5]

plt.pie(portion, explode=(0,0.1,0,0), labels=pets, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('총 가구 대비 반려동물 기르는 비율(2018)')

plt.show()
```
